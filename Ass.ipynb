```python
from pyspark.sql import SparkSession

# Create a SparkSession
spark = SparkSession.builder.appName("YouTubeCategoryClassification").getOrCreate()

# Load the dataset into a Spark DataFrame
spark_df = spark.read.csv("/content/youtube_data.csv", header=True, inferSchema=True)

# Display the schema and some sample data
spark_df.printSchema()
spark_df.show(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the category counts
category_counts = spark_df.groupBy("category").count().orderBy("count", ascending=False).collect()

# Extract categories and counts for plotting
categories = [row['category'] for row in category_counts]
counts = [row['count'] for row in category_counts]

# Create a countplot
plt.figure(figsize=(12, 6))
sns.barplot(x=counts, y=categories)
plt.title('Distribution of Video Categories')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
```

```python
from pyspark.sql.functions import col

# Handle missing values in 'description' and 'hashtags' by filling with empty strings
spark_df = spark_df.fillna("", subset=['description', 'hashtags'])

# Cast relevant columns to appropriate numeric types.
# We'll use a list of columns we want to cast.
numeric_cols = ["duration", "bitrate", "bitrate(video)", "height", "width", "frame rate", "`frame rate(est.)`", "views", "likes", "comments"]

for col_name in numeric_cols:
    spark_df = spark_df.withColumn(col_name, col(col_name).cast("double"))

# Display the schema to confirm the type changes
spark_df.printSchema()
spark_df.show(5)
```

```python
from pyspark.ml.feature import Tokenizer, HashingTF, IDF, VectorAssembler
from pyspark.sql.functions import col, concat_ws

# Combine text columns into a single column
spark_df = spark_df.withColumn("text_features", concat_ws(" ", col("title"), col("description"), col("hashtags")))

# Tokenize the text
tokenizer = Tokenizer(inputCol="text_features", outputCol="words")
words_data = tokenizer.transform(spark_df)

# Apply HashingTF to create term frequency vectors
hashingTF = HashingTF(inputCol="words", outputCol="raw_features", numFeatures=10000)
tf_data = hashingTF.transform(words_data)

# Apply IDF to compute IDF scores and get the final feature vectors
idf = IDF(inputCol="raw_features", outputCol="tfidf_features")
idf_model = idf.fit(tf_data)
tfidf_data = idf_model.transform(tf_data)

# Display the results
tfidf_data.select("title", "text_features", "words", "raw_features", "tfidf_features").show(5)
```

```python
from pyspark.ml.feature import StringIndexer, VectorAssembler
from pyspark.ml.classification import NaiveBayes
from pyspark.ml import Pipeline

# Index the 'category' column
indexer = StringIndexer(inputCol="category", outputCol="label", handleInvalid="skip")

# Assemble the features vector. For now, we'll just use the tfidf_features.
# We can add other numerical features here later if we want to include them.
assembler = VectorAssembler(inputCols=["tfidf_features"], outputCol="features")

# Create a Naive Bayes model
nb = NaiveBayes(smoothing=1.0)

# Create a pipeline to chain the steps
pipeline = Pipeline(stages=[indexer, assembler, nb])

# Split the data into training and testing sets
(trainingData, testingData) = tfidf_data.randomSplit([0.8, 0.2], seed=1234)

# Train the model
model = pipeline.fit(trainingData)

# Make predictions on the test data
predictions = model.transform(testingData)

# Show some of the predictions
predictions.select("category", "label", "probability", "prediction").show(5)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Select (prediction, true label) pairs and compute evaluation metrics.
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print(f"Accuracy = {accuracy}")

# We can also compute other metrics like F1-score, precision, and recall.
evaluator.setMetricName("f1")
f1_score = evaluator.evaluate(predictions)
print(f"F1 Score = {f1_score}")

# Note: Precision and Recall are calculated per class for multiclass classification.
# We can calculate weighted precision and recall.
evaluator.setMetricName("weightedPrecision")
weighted_precision = evaluator.evaluate(predictions)
print(f"Weighted Precision = {weighted_precision}")

evaluator.setMetricName("weightedRecall")
weighted_recall = evaluator.evaluate(predictions)
print(f"Weighted Recall = {weighted_recall}")
```

```python
from pyspark.ml.classification import LogisticRegression

# Create a Logistic Regression model
lr = LogisticRegression(maxIter=10, regParam=0.3, elasticNetParam=0.8)

# Create a pipeline to chain the steps
# We'll reuse the indexer and assembler from the previous Naive Bayes pipeline
pipeline_lr = Pipeline(stages=[indexer, assembler, lr])

# Train the Logistic Regression model
model_lr = pipeline_lr.fit(trainingData)

# Make predictions on the test data
predictions_lr = model_lr.transform(testingData)

# Show some of the predictions
predictions_lr.select("category", "label", "probability", "prediction").show(5)

# Evaluate the Logistic Regression model
evaluator_lr = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
accuracy_lr = evaluator_lr.evaluate(predictions_lr)
print(f"Logistic Regression Accuracy = {accuracy_lr}")

evaluator_lr.setMetricName("f1")
f1_score_lr = evaluator_lr.evaluate(predictions_lr)
print(f"Logistic Regression F1 Score = {f1_score_lr}")

evaluator_lr.setMetricName("weightedPrecision")
weighted_precision_lr = evaluator_lr.evaluate(predictions_lr)
print(f"Logistic Regression Weighted Precision = {weighted_precision_lr}")

evaluator_lr.setMetricName("weightedRecall")
weighted_recall_lr = evaluator_lr.evaluate(predictions_lr)
print(f"Logistic Regression Weighted Recall = {weighted_recall_lr}")
```
